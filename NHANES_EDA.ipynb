```python
# Librerías
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Montamos Google Drive en Colab
from google.colab import drive
drive.mount('/content/drive')

RUTA_BASE = "/content/drive/MyDrive/NHANES/"
```

Cargamos los datasets de NAHNES

```python
# Datos demográficos
demo = pd.read_sas("DEMO_L.xpt.txt")
demo = demo[[
    "SEQN",
    "RIDAGEYR",     # Age in years at screening (artículo)
    "RIAGENDR",     # Gender (artículo)
    "RIDRETH3",     # Race/Hispanic origin w/ NH Asian (artículo)
    "DMDEDUC2",     # Education level - Adults 20+
    "DMDMARTZ",     # Marital status
    'DMDHHSIZ',     # Total number of people in the Household
    "INDFMPIR",     # Ratio of family income to poverty
    "DMQMILIZ",     # Served active duty in US Armed Forces
]]

# Medidas corporales
body_mea = pd.read_sas("BMX_L.xpt.txt")
body_mea = body_mea[[
    "SEQN",
    "BMXHT", #Standing height (cm) (artículo)
    "BMXWT", #Weight (kg) (artículo)
    "BMXBMI", #Body Mass Index (kg/m**2)
    "BMXWAIST" #Waist Circumference (cm)
]]

# Estilo de vida
smoke = pd.read_sas("SMQ_L.xpt.txt")
smoke = smoke[[
    "SEQN",
    "SMQ020", # Smoked at least 100 cigarettes in life
]]

alcohol = pd.read_sas("ALQ_L.xpt.txt")
alcohol = alcohol [[
    "SEQN",
    "ALQ121" #Past 12 mos how often drink alc bev

]]

act_fisica = pd.read_sas("PAQ_L.xpt.txt")
act_fisica = act_fisica[[
    "SEQN",
    "PAD800", #Minutes moderate LTPA
    "PAD680"  #Minutes sedentary activity
]]

sueño = pd.read_sas("SLQ_L.xpt.txt")
sueño = sueño[[
    "SEQN",
    "SLD012" #Sleep hours weekdays or workdays

]]

# Presión arterial
blood_preas = pd.read_sas("BPXO_L.xpt.txt")

blood_preas["SBP_mean"] = blood_preas[["BPXOSY1","BPXOSY2","BPXOSY3"]].mean(axis=1)
blood_preas["DBP_mean"] = blood_preas[["BPXODI1","BPXODI2","BPXODI3"]].mean(axis=1)
blood_preas = blood_preas[[
    "SEQN",
    "SBP_mean", #Media de mediciones (3) de presión sistólica
    "DBP_mean" #Media de mediciones (3) de presión diastólica
]]

# Datos de laboratorio
chol = pd.read_sas("TCHOL_L.xpt.txt")
chol = chol[[
    "SEQN",
    "LBXTC" #Total cholesterol (mg/dl)
]]

hdl = pd.read_sas("HDL_L.xpt.txt")
hdl = hdl[[
    "SEQN",
    "LBDHDD" #Direct HDL-Cholesterol (mg/dl)
]]


glicoHb = pd.read_sas("GHB_L.xpt.txt")
glicoHb = glicoHb[[
    "SEQN",
    "LBXGH" #Glicohemoglobina (%)
]]

prot_C_react = pd.read_sas("HSCRP_L.xpt.txt")
prot_C_react = prot_C_react[[
    "SEQN",
    "LBXHSCRP" #HS C-Reactive Protein (mg/dl)
]]

bioq = pd.read_sas("BIOPRO_L.xpt.txt")
bioq = bioq[[
    "SEQN",
    "LBXSCR", #Creatinina en Suero (mg/dl)
    "LBXSBU", #Blood Urea Nitrogen (mg/dl)
    "LBXSUA"  #Uric acid (mg/dl)
]]

blood = pd.read_sas("CBC_L.xpt.txt")
blood = blood[[
    "SEQN",
    "LBXHGB", #Hemoglobin (g/dL)
    "LBXWBCSI", #White blood cell count (1000 cells/uL)
    "LBXRBCSI", #Red blood cell count (million cells/uL)
]]

# Variables psicológicas
anim = pd.read_sas("DPQ_L.xpt.txt")
anim = anim[[
    "SEQN",
    "DPQ020", #Feeling down, depressed, or hopeless
]]

# Seguro médico
insurance = pd.read_sas("HIQ_L.xpt.txt")
insurance = insurance[[
    "SEQN",
    "HIQ011" #Covered by health insurance
]]

# Salud general
healt_cond = pd.read_sas("HUQ_L.xpt.txt")
healt_cond = healt_cond[[
    "SEQN",
    "HUQ010", #General health condition
    "HUQ030" #Routine place to go for healthcare
]]


# Historial médico
hypert_chol = pd.read_sas("BPQ_L.xpt.txt")
hypert_chol = hypert_chol[[
    "SEQN",
    "BPQ020", #Doctor told you have hypertension (artículo)
    "BPQ080", #Doctor told you have high cholesterol level
    "BPQ101D", #Taking meds to lower blood cholesterol?
]]

diabetes = pd.read_sas("DIQ_L.xpt.txt")
diabetes = diabetes[[
    "SEQN",
    "DIQ010" #Doctor told you have diabetes (artículo)
]]

med = pd.read_sas("MCQ_L.xpt.txt")
med = med[[
    "SEQN",
    "MCQ160A",  # Doctor ever said you had arthritis (artículo)
    "MCQ160B",  # Ever told had congestive heart failure (artículo)
    "MCQ160C",  # Ever told you had coronary heart desease (artículo)
    "MCQ160D",  # Ever told you had angina/angina pectoris (artículo)
    "MCQ160E",  # Ever told you had heart attack (artículo)
    "MCQ160P",  # Ever told you had COPD, emphysema, ChB (artículo)
    "MCQ160L",  # Ever told you had any liver condition (artículo)
    "MCQ220",   # Ever told you had cancer or malignancy (artículo)
    "MCQ160M",  # Ever told you had thyroid problem
    "MCQ160F",  # Ever told you had a stroke (artículo)
]]
```

Número de participantes en cada archivo

```python
# Longitud de cada archivo
data_lengths = {
    "demo": len(demo),
    "body_mea": len(body_mea),
    "smoke": len(smoke),
    "alcohol": len(alcohol),
    "act_fisica": len(act_fisica),
    "sueño": len(sueño),
    "blood_preas": len(blood_preas),
    "chol": len(chol),
    "hdl": len(hdl),
    "glicoHb": len(glicoHb),
    "prot_C_react": len(prot_C_react),
    "bioq": len(bioq),
    "blood": len(blood),
    "anim": len(anim),
    "insurance": len(insurance),
    "healt_cond": len(healt_cond),
    "hypert_chol": len(hypert_chol),
    "diabetes": len(diabetes),
    "med": len(med)
}

lengths_df = pd.DataFrame(list(data_lengths.items()), columns=['Dataset', 'Número de Participantes'])
print(lengths_df.to_string(index=False))
```

Unimos toda la información disponible de los participantes a partir del dataset demográfico `demo`.



```python
# Merge completo de todos los datasets seleccionados
df = demo

dfs = [
    body_mea, smoke, alcohol, act_fisica, sueño,
    blood_preas, chol, hdl, glicoHb, prot_C_react,
    bioq, blood, anim, insurance, healt_cond,
    hypert_chol, diabetes, med
]

for d in dfs:
    df = df.merge(d, on="SEQN", how="left")

# Dimensiones antes de eliminar los pacientes NaN para la variable objetivo.
df.shape
```

Renombramos las variables del dataset

```python
#Definimos el nombre de las variables
rename_dict = {
    # Datos demográficos
    "SEQN": "ID",
    "RIDAGEYR": "Edad",
    "RIAGENDR": "Sexo",
    "RIDRETH3": "Raza",
    "DMDEDUC2": "Educacion",
    "DMDMARTZ": "Estado_civil",
    "DMDHHSIZ": "Personas_hogar",
    "INDFMPIR": "Ingresos",
    "DMQMILIZ": "Militar",

    # Medidas corporales
    "BMXHT": "Altura",
    "BMXWT": "Peso",
    "BMXBMI": "IMC",
    "BMXWAIST": "Cintura",

    # Presión arterial
    "SBP_mean": "PAS_media",
    "DBP_mean": "PAD_media",

    # Laboratorio - Lípidos
    "LBXTC": "Colesterol_total",
    "LBDHDD": "HDL_directo",

    # Laboratorio - Glucosa
    "LBXGH": "HbA1c",

    # Laboratorio - Inflamación / riñón
    "LBXHSCRP": "Proteina_C_reactiva",
    "LBXSCR": "Creatinina",
    "LBXSBU": "BUN",
    "LBXSUA": "Acido_urico",

    # Laboratorio - Recuentos en sangre
    "LBXHGB": "Hemoglobina",
    "LBXWBCSI": "Leucocitos",
    "LBXRBCSI": "Globulos_rojos",

    # Estilo de vida
    "SMQ020": "Fumador_100cigs",
    "ALQ121": "Alcohol_12meses",
    "PAD800": "Minutos_LTPA_moderada",
    "PAD680": "Minutos_sedentario",

    # Sueño
    "SLD012": "Tiempo_sueño",

    # Psicología
    "DPQ020": "Sentimiento_deprimido",

    # Acceso a servicios de salud / Seguro médico
    "HIQ011": "Seguro_medico",
    "HUQ010": "Estado_salud_general",
    "HUQ030": "Acceso_atencion_sanitaria",

    # Historial médico
    "BPQ020": "Hipertension",
    "BPQ080": "Colesterol_alto",
    "BPQ101D": "Medicacion_colesterol",
    "DIQ010": "Diabetes",

    # Condición médica
    "MCQ160A": "Artritis",
    "MCQ160B": "Insuficiencia_cardiaca",
    "MCQ160C": "Enfermedad_coronaria",
    "MCQ160D": "Angina",
    "MCQ160E": "Infarto_miocardio",
    "MCQ160P": "Enfisema_ChB",
    "MCQ160L": "Enfermedad_hepatica",
    "MCQ220": "Cancer",
    "MCQ160M": "Problema_tiroides",
    "MCQ160F": "Stroke"
}

#Aplicamos el renombrado
df = df.rename(columns=rename_dict)
```

```python
# Códigos especiales tratados como NaN
df["Stroke"] = df["Stroke"].replace([7, 9], np.nan)

# Participantes con información válida para Stroke
df = df[df["Stroke"].notna()]
```

## Transformación de variables

```python
# Variables categóricas
var_categoricas = [
    'Sexo','Raza','Educacion','Estado_civil', 'Militar', 'Fumador_100cigs',
    'Alcohol_12meses','Sentimiento_deprimido','Seguro_medico',
    'Estado_salud_general','Acceso_atencion_sanitaria','Hipertension',
    'Colesterol_alto','Medicacion_colesterol','Diabetes',
    'Artritis','Insuficiencia_cardiaca','Enfermedad_coronaria',
    'Angina','Infarto_miocardio','Enfisema_ChB','Enfermedad_hepatica',
    'Cancer','Problema_tiroides'
]

# Variables numéricas
var_numericas = [col for col in df.columns if col not in var_categoricas + ['Stroke'] + ['ID']]
```

### Variable objetivo

```python
df['Stroke'] = df['Stroke'].map({
    1.0: 'Sí',
    2.0: 'No'
}).astype('category')
```

### Variables categóricas

Tratamos aquellos casos en los que rechazan contestar o desconocen la respuesta como valores faltantes y reconvertimos algunas de ellas a oridinales continuas.

```python
# Redondeo de variables categóricas (para evitar problemas con 0.0)
df[var_categoricas] = df[var_categoricas].round(0)

# Sexo
df['Sexo'] = df['Sexo'].map({1.0: 'Hombre', 2.0: 'Mujer', 7.0: np.nan}).astype('category')

# Raza / origen étnico
df['Raza'] = df['Raza'].map({
    1.0: 'Hispano',
    2.0: 'Hispano',
    3.0: 'Blanco',
    4.0: 'Negro',
    6.0: 'Asiático',
    7.0: 'Otro',
}).astype('category')

# Educación –> ordinal continua
df['Educacion'] = df['Educacion'].map({
    1.0: 1, # <9º curso
    2.0: 2, # 9º-11º
    3.0: 3, # Secundaria
    4.0: 4, # Universidad empezada
    5.0: 5, # Universitario o más
    7.0: np.nan,
    9.0: np.nan
})

# Estado civil
df['Estado_civil'] = df['Estado_civil'].map({
    1.0: 'Pareja',
    2.0: 'Separado/Viudo',
    3.0: 'Soltero',
    77.0: np.nan,
    99.0: np.nan
}).astype('category')

# Servicio militar
df['Militar'] = df['Militar'].map({
    1.0: 'Sí',
    2.0: 'No',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Fumador
df['Fumador_100cigs'] = df['Fumador_100cigs'].map({
    1.0: 'Sí',
    2.0: 'No',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Alcohol últimos 12 meses -> ordinal continua
df['Alcohol_12meses'] = df['Alcohol_12meses'].map({
    0.0: 0, #Nunca
  10.0: 1, 9.0: 1, 8.0: 1, 7.0: 1,     # Bajo
    6.0: 2, 5.0: 2,                    # Moderado
    4.0: 3, 3.0: 3,                    # Alto
    2.0: 4, 1.0: 4,                    # Muy alto
    77.0: np.nan,
    99.0: np.nan
    })

# Depresión –> ordinal continua
df['Sentimiento_deprimido'] = df['Sentimiento_deprimido'].map({
    0.0: 0,
    1.0: 1,
    2.0: 2,
    3.0: 3,
    7.0: np.nan,
    9.0: np.nan
})

# Seguro médico
df['Seguro_medico'] = df['Seguro_medico'].map({
    1.0: 'Sí',
    2.0: 'No',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Estado salud general –> ordinal continua
df['Estado_salud_general'] = df['Estado_salud_general'].map({
    1.0: 5, # Excelente
    2.0: 4, # Muy buena
    3.0: 3, # Buena
    4.0: 2, # Regular
    5.0: 1, # Mala
    7.0: np.nan,
    9.0: np.nan
})

# Acceso atención sanitaria –> simplificar
df['Acceso_atencion_sanitaria'] = df['Acceso_atencion_sanitaria'].map({
    1.0: 'Sí',
    2.0: 'No',
    3.0: 'Sí',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Hipertensión
df['Hipertension'] = df['Hipertension'].map({
    1.0: 'Sí',
    2.0: 'No',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Colesterol alto
df['Colesterol_alto'] = df['Colesterol_alto'].map({
    1.0: 'Sí',
    2.0: 'No',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Medicación colesterol
df['Medicacion_colesterol'] = df['Medicacion_colesterol'].map({
    1.0: 'Sí',
    2.0: 'No',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Diabetes –> ordinal
df['Diabetes'] = df['Diabetes'].map({
    1.0: 'Sí',
    2.0: 'No',
    3.0: 'Pre-diabetes',
    7.0: np.nan,
    9.0: np.nan
}).astype('category')

# Condiciones médicas
categorical_conditions = [
    'Artritis', 'Insuficiencia_cardiaca', 'Enfermedad_coronaria',
    'Angina', 'Infarto_miocardio', 'Enfisema_ChB', 'Enfermedad_hepatica',
    'Cancer', 'Problema_tiroides'
]

for col in categorical_conditions:
    df[col] = df[col].map({
        1.0: 'Sí',
        2.0: 'No',
        7.0: np.nan,
        9.0: np.nan
    }).astype('category')
```

### Variables numéricas

Tratamos aquellos casos en los que rechazan contestar o desconocen la respuesta como valores faltantes

```python
# Tratamiento de valores faltantes en variables numéricas
replacements = {
    "Minutos_LTPA_moderada": [7777, 9999],
    "Minutos_sedentario": [7777, 9999],
}

for col, values_to_replace in replacements.items():
    df[col] = df[col].replace(values_to_replace, np.nan)
```

Actualizamos el tipo de variables, ya que algunas categóricas se han converitdo a ordinales numéricas

```python
# Variables categóricas
var_categoricas = [
    'Sexo','Raza','Estado_civil','Militar','Fumador_100cigs',
    'Seguro_medico','Acceso_atencion_sanitaria','Hipertension',
    'Colesterol_alto','Medicacion_colesterol','Diabetes',
    'Artritis','Insuficiencia_cardiaca','Enfermedad_coronaria',
    'Angina','Infarto_miocardio','Enfisema_ChB','Enfermedad_hepatica',
    'Cancer','Problema_tiroides'
]

# Variables ordinales (convertidas a numéricas)
var_ordinales = [
    'Educacion',
    'Alcohol_12meses',
    'Sentimiento_deprimido',
    'Estado_salud_general'
]

# Variables numéricas
var_numericas = [
    col for col in df.columns
    if col not in var_categoricas + var_ordinales + ['Stroke','ID']
]

# Variables ordinales + variables numéricas
var_numericas_total = var_ordinales + var_numericas
```

## Valores faltantes

Cálculo del número de Nan por columna

```python
# Valores faltantes por columna
nan_por_columna = df.isna().sum().sort_values(ascending=False)

# Porcentaje de NaN por columna
porcentaje_nan_por_columna = df.isna().mean().sort_values(ascending=False) * 100

nan_columna_df = pd.DataFrame({
    'Num_NaN_Columna': nan_por_columna,
    'Porcentaje_Participantes': porcentaje_nan_por_columna.round(2)
    })
print(nan_columna_df)
```

```python
# Visualización de valores faltantes por columna
nan_columna_df_plot = nan_columna_df[nan_columna_df['Num_NaN_Columna'] > 0]

plt.figure(figsize=(10,5))
plt.bar(nan_columna_df_plot.index, nan_columna_df_plot['Porcentaje_Participantes'])
plt.xticks(rotation=45, ha='right')
plt.ylabel('% de participantes con NaN')
plt.title('Valores faltantes por columna')
plt.tight_layout()
plt.show()
```

Cálculo del número de NaN por fila

```python
# Valores faltantes por fila
nan_fila = df.isna().sum(axis=1).value_counts()
nan_fila_percent = df.isna().sum(axis=1).value_counts(normalize=True).sort_index() * 100

# Porcentaje de pacientes para cada NaN
nan_fila_percent = df.isna().sum(axis=1).value_counts(normalize=True).sort_index() * 100

nan_fila_df = pd.DataFrame({
    'Num_Obs_por_NaN': nan_fila.index,
    'Porcentaje_Participantes': nan_fila_percent.values
})
print(nan_fila_df)
```

```python
# Visualización
plt.figure(figsize=(10,5))
plt.bar(nan_fila_df['Num_Obs_por_NaN'], nan_fila_df['Porcentaje_Participantes'])
plt.xlabel('Número de observaciones para cada valor de NaN')
plt.ylabel('% de participantes')
plt.title('Distribución de valores faltantes por fila')
plt.xticks(rotation=0)
plt.show()
```

#### Correlación variables numéricas

```python
# Variables agrupadas por bloque conceptual
bloques = {
    "Sociodemográfico":        ["Edad", "Ingresos", "Educacion", "Personas_hogar"],
    "Estilo de vida y mental": ["Estado_salud_general", "Sentimiento_deprimido", "Alcohol_12meses",
                                "Tiempo_sueño", "Minutos_LTPA_moderada", "Minutos_sedentario"],
    "Antropométrico":          ["Altura", "Peso", "IMC", "Cintura"],
    "Presión arterial":        ["PAS_media", "PAD_media"],
    "Lipídico":                ["Colesterol_total", "HDL_directo"],
    "Glucémico":               ["HbA1c"],
    "Renal y metabólico":      ["Creatinina", "BUN", "Acido_urico"],
    "Hematológico":            ["Hemoglobina", "Globulos_rojos", "Leucocitos"],
    "Inflamación":             ["Proteina_C_reactiva"],
}
orden = [v for grupo in bloques.values() for v in grupo]

corr = df[orden].corr(method="pearson")

fig, ax = plt.subplots(figsize=(12, 11))
sns.heatmap(corr, cmap="coolwarm", center=0, vmin=-1, vmax=1, square=True,
            linewidths=0.4, linecolor="white",
            cbar_kws={"label": "Coeficiente de correlación de Pearson", "shrink": 0.8}, ax=ax)


ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=8)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=8)
plt.tight_layout()
plt.savefig(RUTA_BASE + "figura_anexo_corr_nhanes.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Variables categóricas vs Stroke

```python
# Colores azul y naranja
colors = ['#ff7f0e', '#1f77b4']  # azul para Stroke Sí, naranja para Stroke No

for col in var_categoricas:
    if col != "Stroke":
        # Tabla de porcentajes por fila (cada categoría suma 100%)
        tabla = pd.crosstab(df[col], df["Stroke"], normalize='index') * 100
        tabla = tabla[['Sí', 'No']]  # Aseguramos mismo orden de columnas
        print(f"\n{col} vs Stroke (%)")
        print(tabla.round(2))

        # Gráfico de barras apiladas usando la misma tabla
        ax = tabla.plot(kind='bar', stacked=True, figsize=(6,4), color=colors)
        plt.title(f"{col} vs Stroke (%)")
        plt.ylabel("Proporción de Stroke en cada grupo")
        plt.xlabel(col)
        plt.xticks(rotation=45)
        plt.ylim(0,100)
        plt.legend(title='Stroke')
        plt.tight_layout()
        plt.show()
```
